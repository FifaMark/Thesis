1. Imports & Global Settings

```python
import copy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor  # kept for compatibility
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, RegressorMixin

# River: online Hoeffding Tree
from river.tree import HoeffdingTreeRegressor as RiverHTRegressor

# Keep warnings/style exactly as before
warnings.filterwarnings('ignore', category=FutureWarning)
plt.style.use('seaborn-v0_8-whitegrid')

# --- sklearn-compatible wrapper around a River regressor (clone-safe) ---
class RiverRegressor(BaseEstimator, RegressorMixin):
    """
    Wrap a river regressor for use in an sklearn Pipeline.
    Prefer passing (river_model_cls, river_model_kwargs) so we can instantiate
    a fresh model inside fit(), which is clone-safe for sklearn.
    """
    def __init__(self, river_model=None, river_model_cls=RiverHTRegressor, river_model_kwargs=None):
        self.river_model = river_model                  # optional instance (not recommended)
        self.river_model_cls = river_model_cls          # preferred: class
        self.river_model_kwargs = river_model_kwargs or {}

    def _fresh_model(self):
        if self.river_model is not None:
            return copy.deepcopy(self.river_model)
        cls = self.river_model_cls or RiverHTRegressor
        return cls(**self.river_model_kwargs)

    def fit(self, X, y):
        model = self._fresh_model()
        X = np.asarray(X)
        y = np.asarray(y).ravel()
        for xi, yi in zip(X, y):
            x_dict = {i: float(v) for i, v in enumerate(np.asarray(xi).ravel())}
            model.learn_one(x_dict, float(yi))
        self.model_ = model
        return self

    def predict(self, X):
        X = np.asarray(X)
        preds = []
        model = getattr(self, "model_", None)
        if model is None:
            model = self._fresh_model()
        for xi in X:
            x_dict = {i: float(v) for i, v in enumerate(np.asarray(xi).ravel())}
            yhat = model.predict_one(x_dict)
            preds.append(float(0.0 if yhat is None else yhat))
        return np.array(preds, dtype=float)
```

2. Data Loading & Target Definition


```python
# 2.1 Specify columns to read
usecols = [
    'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'Referee',
    'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR',
    'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5'
]

file_path = r"D:\python\premier league predictor\premier league data\premier leaguer 22-23.csv"
df = pd.read_csv(file_path, usecols=usecols)

# 2.2 Assign a sequential MatchIndex and compute Round number
df['MatchIndex'] = df.index + 1
df['Round'] = np.ceil(df['MatchIndex'] / 10).astype(int)

# 2.3 Extract hour from kickoff Time and bucket into Morning/Afternoon/Evening
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.hour
bins = [-1, 11, 17, 24]
labels = ['Morning', 'Afternoon', 'Evening']
df['TimeOfDay'] = pd.cut(df['Hour'], bins=bins, labels=labels)
df.drop(columns='Hour', inplace=True)

# 2.4 Define the regression target
df['TotalYellows'] = df['HY'] + df['AY']
df['LogYellows'] = np.log1p(df['TotalYellows'])

print("Loaded shape:", df.shape)
df.head()
```

3. Leakage-Free Feature Engineering

```python
# 3.1 Map full-time result to numeric Outcome (Home=1, Draw=0, Away=-1)
df['Outcome'] = df['FTR'].map({'H': 1, 'D': 0, 'A': -1})

# 3.2 Implied probabilities from Bet365 odds (normalized per row)
for side in ['H', 'D', 'A']:
    df[f'ImpProb_{side}'] = 1.0 / df[f'B365{side}']
prob_cols = ['ImpProb_H', 'ImpProb_D', 'ImpProb_A']
df[prob_cols] = df[prob_cols].div(df[prob_cols].sum(axis=1), axis=0)

# Over/Under 2.5 (normalized per row)
df['ImpOver2.5'] = 1.0 / df['B365>2.5']
df['ImpUnder2.5'] = 1.0 / df['B365<2.5']
df[['ImpOver2.5', 'ImpUnder2.5']] = (
    df[['ImpOver2.5', 'ImpUnder2.5']]
      .div(df[['ImpOver2.5', 'ImpUnder2.5']].sum(axis=1), axis=0)
)

# 3.3 Odds spread (Home minus Away)
df['OddsSpread'] = df['B365H'] - df['B365A']

# 3.4 Rolling 3-match form (shifted so only past games are used)
df['HomeForm3'] = (
    df.groupby('HomeTeam')['Outcome']
      .transform(lambda s: s.shift(1).rolling(3, min_periods=1).mean())
)
df['AwayForm3'] = (
    df.groupby('AwayTeam')['Outcome']
      .transform(lambda s: s.shift(1).rolling(3, min_periods=1).mean())
)

# 3.5 Win/Loss streaks (shifted)
df['HomeStreak'] = (
    df.groupby('HomeTeam')['Outcome']
      .transform(lambda s: s.eq(1).shift(1).fillna(0).astype(int)
                    .groupby((~s.eq(1)).cumsum()).cumsum())
)
df['AwayStreak'] = (
    df.groupby('AwayTeam')['Outcome']
      .transform(lambda s: s.eq(-1).shift(1).fillna(0).astype(int)
                    .groupby((~s.eq(-1)).cumsum()).cumsum())
)

# 3.6 Interaction term: odds spread × home form
df['Spread×Form'] = df['OddsSpread'] * df['HomeForm3']

# 3.7 Referee home-bias (shifted)
df['RefHomeWins'] = (
    df.groupby('Referee')['Outcome']
      .transform(lambda s: s.eq(1).cumsum().shift(1).fillna(0))
)
df['RefMatches'] = df.groupby('Referee').cumcount()
df['RefBias'] = (df['RefHomeWins'] / df['RefMatches']).fillna(0.5)
df.drop(columns=['RefHomeWins', 'RefMatches'], inplace=True)

# 3.8 Head-to-head last result
df['PairKey'] = df.apply(lambda r: tuple(sorted([r.HomeTeam, r.AwayTeam])), axis=1)
h2h = {}
def prev_h2h(row):
    key = row.PairKey
    prev = h2h.get(key, 0)
    sign = 1 if row.HomeTeam == key[0] else -1
    h2h[key] = row.Outcome * sign
    return prev
df['H2H_Last'] = df.apply(prev_h2h, axis=1)
df.drop(columns='PairKey', inplace=True)

# 3.9 Rolling averages of match stats
home_stats = ['HS', 'HST', 'HF', 'HC', 'HY', 'HR']
away_stats = ['AS', 'AST', 'AF', 'AC', 'AY', 'AR']

for stat in home_stats:
    df[f'Home_{stat}_3'] = (
        df.groupby('HomeTeam')[stat]
          .transform(lambda s: s.shift(1).rolling(3, min_periods=1).mean())
    )

for stat in away_stats:
    df[f'Away_{stat}_3'] = (
        df.groupby('AwayTeam')[stat]
          .transform(lambda s: s.shift(1).rolling(3, min_periods=1).mean())
    )

# 3.10 Previous-round table positions
df['HomePts'] = np.where(df['FTR'] == 'H', 3,
                         np.where(df['FTR'] == 'D', 1, 0))
df['AwayPts'] = np.where(df['FTR'] == 'A', 3,
                         np.where(df['FTR'] == 'D', 1, 0))

pts = pd.DataFrame({
    'Team': np.concatenate([df['HomeTeam'], df['AwayTeam']]),
    'Round': np.concatenate([df['Round'], df['Round']]),
    'Pts': np.concatenate([df['HomePts'], df['AwayPts']])
})

round_pts = (
    pts.groupby(['Team', 'Round'], as_index=False)['Pts']
       .sum()
)
round_pts['CumPts'] = round_pts.groupby('Team')['Pts'].cumsum()
round_pts['Pos'] = (
    round_pts.groupby('Round')['CumPts']
             .rank(method='min', ascending=False)
)

prev = round_pts[['Team', 'Round', 'Pos']].copy()
prev.rename(columns={'Round': 'PrevRound', 'Pos': 'PrevPos'}, inplace=True)
prev['PrevRound'] += 1

df = df.merge(prev,
              left_on=['HomeTeam', 'Round'],
              right_on=['Team', 'PrevRound'],
              how='left')
df['HomePosPrev'] = df['PrevPos'].fillna(df['Round'].max() + 1)
df.drop(columns=['Team', 'PrevRound', 'PrevPos'], inplace=True)

df = df.merge(prev,
              left_on=['AwayTeam', 'Round'],
              right_on=['Team', 'PrevRound'],
              how='left')
df['AwayPosPrev'] = df['PrevPos'].fillna(df['Round'].max() + 1)
df.drop(columns=['Team', 'PrevRound', 'PrevPos'], inplace=True)

# 3.11 Position difference
df['PosDiff'] = df['HomePosPrev'] - df['AwayPosPrev']
```

4. Pipeline & Feature Schema (Polynomial + Hoeffding Tree Regressor)

```python
# 4.1 Categorical and numeric feature lists
cat_feats = ['HomeTeam', 'AwayTeam', 'Referee', 'TimeOfDay']

num_feats = [
    'HomeForm3', 'AwayForm3',
    'HomeStreak', 'AwayStreak',
    'ImpProb_H', 'ImpProb_D', 'ImpProb_A',
    'ImpOver2.5', 'ImpUnder2.5',
    'OddsSpread', 'Spread×Form',
    'RefBias', 'H2H_Last',
    'HomePosPrev', 'AwayPosPrev',
    'PosDiff'
] + [f'Home_{stat}_3'] * 0  # placeholder to keep structure stable

# Extend with rolling stats from §3.9
num_feats += [f'Home_{stat}_3' for stat in ['HS','HST','HF','HC','HY','HR']]
num_feats += [f'Away_{stat}_3' for stat in ['AS','AST','AF','AC','AY','AR']]

# 4.2 Fill NaNs in numeric features
df[num_feats] = df[num_feats].fillna(0)
df['TimeOfDay'] = df['TimeOfDay'].astype(object).fillna('Afternoon')

# 4.3 Preprocessor with PolynomialFeatures for numeric branch
from sklearn.preprocessing import PolynomialFeatures, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

def make_preprocessor_poly(degree=5, interaction_only=False, include_bias=False):
    num_pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('poly', PolynomialFeatures(degree=degree,
                                    interaction_only=interaction_only,
                                    include_bias=include_bias))
    ])
    cat_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    return ColumnTransformer([
        ('num', num_pipeline, num_feats),
        ('cat', cat_encoder,  cat_feats),
    ], remainder='drop')

# 4.4 Model factories — River Hoeffding Tree (sklearn-compatible via wrapper)
from river.tree import HoeffdingTreeRegressor as RiverHTRegressor
from sklearn.compose import TransformedTargetRegressor
import numpy as np

def make_model_ht(degree=2, interaction_only=False, **ht_kwargs):
    """
    Polynomial features + River Hoeffding Tree on original target scale.
    Default leaf_prediction='mean' for stability (no extrapolation).
    """
    pre = make_preprocessor_poly(degree=degree, interaction_only=interaction_only)
    # defaults for river
    kwargs = dict(grace_period=200, delta=1e-7, tau=0.05, leaf_prediction='mean')
    kwargs.update(ht_kwargs or {})
    return Pipeline([
        ('prep', pre),
        ('reg',  RiverRegressor(river_model_cls=RiverHTRegressor, river_model_kwargs=kwargs))
    ])

def _safe_expm1_cap(z, cap=40.0):
    """Safer inverse for log1p; clips in log-space, returns non-negative outputs."""
    z = np.asarray(z, dtype=float)
    z = np.clip(z, -50.0, np.log1p(cap))
    out = np.expm1(z)
    return np.clip(out, 0.0, cap)

def make_model_ht_log(degree=2, interaction_only=False, **ht_kwargs):
    """
    Same as above but with log1p target transform (used only if negatives would occur).
    Uses a safe inverse to avoid overflow.
    """
    pre = make_preprocessor_poly(degree=degree, interaction_only=interaction_only)
    kwargs = dict(grace_period=200, delta=1e-7, tau=0.05, leaf_prediction='mean')
    kwargs.update(ht_kwargs or {})
    log_wrap = TransformedTargetRegressor(
        regressor=RiverRegressor(river_model_cls=RiverHTRegressor, river_model_kwargs=kwargs),
        func=np.log1p,
        inverse_func=lambda z: _safe_expm1_cap(z, cap=40.0)
    )
    return Pipeline([
        ('prep', pre),
        ('reg',  log_wrap)
    ])
```

5. Round-by-Round Training & Prediction (Hoeffding Tree Regressor)

```python
preds = []
preds_rounded = []
actuals = []
rounds = []
match_indices = []

warmup = 50
degree = 2
interaction_only = False

# River-friendly hyperparameters (supported by river)
# Use leaf_prediction='mean' to avoid extrapolation-heavy leaves
ht_kwargs = dict(
    grace_period=300,   # slightly larger to reduce overfitting/splits
    delta=1e-5,         # a bit looser split confidence (fewer fragile splits)
    tau=0.05,
    leaf_prediction='mean'
)

def _winsorize_upper(y_pred, y_train):
    """
    Dynamic, data-driven upper winsorization:
    upper cap = max( (99.5th percentile of train) + 2, 12 )
    Only upper bound; lower bound is 0.
    """
    y_pred = np.asarray(y_pred, dtype=float)
    y_train = np.asarray(y_train, dtype=float)
    if y_train.size > 0:
        hi = float(np.nanpercentile(y_train, 99.5)) + 2.0
        hi = max(hi, 12.0)     # ensure some headroom even if training low-variance
    else:
        hi = 20.0
    y_pred = np.nan_to_num(y_pred, nan=0.0, posinf=hi, neginf=0.0)
    y_pred = np.clip(y_pred, 0.0, hi)
    return y_pred

# 5.1 Initial warm-up fit on the first `warmup` matches
model = make_model_ht(degree=degree, interaction_only=interaction_only, **ht_kwargs)
model.fit(
    df.loc[:warmup-1, cat_feats + num_feats],
    df.loc[:warmup-1, 'TotalYellows']
)

# 5.2 Walk-forward by round (train on past rounds, predict the current round)
for r in range(6, df['Round'].max() + 1):
    train_idx = df['Round'] < r
    test_idx  = df['Round'] == r

    X_train = df.loc[train_idx, cat_feats + num_feats]
    y_train = df.loc[train_idx, 'TotalYellows']

    X_test  = df.loc[test_idx,  cat_feats + num_feats]
    y_test  = df.loc[test_idx,  'TotalYellows']

    # Train non-log model first
    model = make_model_ht(degree=degree, interaction_only=interaction_only, **ht_kwargs)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # If any negative predictions, switch to log model (safe inverse); otherwise keep non-log
    if np.any(y_pred < 0):
        model = make_model_ht_log(degree=degree, interaction_only=interaction_only, **ht_kwargs)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

    # Dynamic winsorization: only 0 floor + train-based upper cap (no fixed 30)
    y_pred = _winsorize_upper(y_pred, y_train)

    # Store predictions and metadata
    preds.extend(y_pred.tolist())
    preds_rounded.extend(np.rint(y_pred).astype(int).tolist())
    actuals.extend(y_test.tolist())
    rounds.extend([r] * len(y_test))
    match_indices.extend(df.loc[test_idx, 'MatchIndex'].tolist())

print(f"Predicted rounds 6 to {df['Round'].max()}")

# 5.3 Keep the final trained model for the Gradio interface
final_model = model
```

```python
import gradio as gr

# 8.1 Precompute mean values for odds-derived features
odd_feats = [
    'ImpProb_H','ImpProb_D','ImpProb_A',
    'ImpOver2.5','ImpUnder2.5','OddsSpread'
]
odd_means = df[odd_feats].mean()

# 8.2 UI dropdown choices
teams = sorted(df['HomeTeam'].unique())
refs  = sorted(df['Referee'].unique())
times = ['Morning','Afternoon','Evening']

# 8.3 Prediction function (raw & rounded)
def predict_yellows(home_team, away_team, referee, time_of_day):
    # 1) Last known rows per entity
    try:
        home_last = df[df.HomeTeam == home_team].iloc[-1]
    except IndexError:
        home_last = pd.Series(dtype=float)

    try:
        away_last = df[df.AwayTeam == away_team].iloc[-1]
    except IndexError:
        away_last = pd.Series(dtype=float)

    try:
        ref_last  = df[df.Referee == referee].iloc[-1]
    except IndexError:
        ref_last = pd.Series(dtype=float)

    # 2) Odds-derived features (imputed with means)
    feat = {
        'ImpProb_H':   odd_means['ImpProb_H'],
        'ImpProb_D':   odd_means['ImpProb_D'],
        'ImpProb_A':   odd_means['ImpProb_A'],
        'ImpOver2.5':  odd_means['ImpOver2.5'],
        'ImpUnder2.5': odd_means['ImpUnder2.5'],
        'OddsSpread':  odd_means['OddsSpread'],
    }

    # 3) History-based features
    feat['HomeForm3']   = float(home_last.get('HomeForm3', 0))
    feat['AwayForm3']   = float(away_last.get('AwayForm3', 0))
    feat['HomeStreak']  = float(home_last.get('HomeStreak', 0))
    feat['AwayStreak']  = float(away_last.get('AwayStreak', 0))
    feat['RefBias']     = float(ref_last.get('RefBias', 0.5))
    feat['HomePosPrev'] = float(home_last.get('HomePosPrev', 0))
    feat['AwayPosPrev'] = float(away_last.get('AwayPosPrev', 0))
    feat['PosDiff']     = feat['HomePosPrev'] - feat['AwayPosPrev']

    # 4) Head-to-head last result signed for the current home
    h2h_mask = ((df['HomeTeam'] == home_team) & (df['AwayTeam'] == away_team)) \
               | ((df['HomeTeam'] == away_team) & (df['AwayTeam'] == home_team))
    if h2h_mask.any():
        last_h2h_row = df[h2h_mask].iloc[-1]
        h2h_val = float(last_h2h_row.get('H2H_Last', 0))
        if last_h2h_row['HomeTeam'] != home_team:
            h2h_val = -h2h_val
        feat['H2H_Last'] = h2h_val
    else:
        feat['H2H_Last'] = 0.0

    # 5) Rolling stat features
    for stat in ['HS','HST','HF','HC','HY','HR']:
        feat[f'Home_{stat}_3'] = float(home_last.get(f'Home_{stat}_3', 0))
    for stat in ['AS','AST','AF','AC','AY','AR']:
        feat[f'Away_{stat}_3'] = float(away_last.get(f'Away_{stat}_3', 0))

    # 6) Interaction term
    feat['Spread×Form'] = feat['OddsSpread'] * feat['HomeForm3']

    # 7) Categorical features
    feat['HomeTeam']  = home_team
    feat['AwayTeam']  = away_team
    feat['Referee']   = referee
    feat['TimeOfDay'] = time_of_day

    # 8) Assemble single-row input with the exact schema
    X_row = pd.DataFrame([{**{k: feat[k] for k in cat_feats},
                           **{k: feat[k] for k in num_feats}}])
    for c in cat_feats:
        X_row[c] = X_row[c].astype(object)
    X_row[num_feats] = X_row[num_feats].fillna(0)

    # 9) Predict with the final walk-forward model (Hoeffding Tree)
    y_pred = float(final_model.predict(X_row)[0])
    y_pred_round = int(np.rint(y_pred))

    return y_pred, y_pred_round

# 8.4 Gradio interface
iface = gr.Interface(
    fn=predict_yellows,
    inputs=[
        gr.Dropdown(choices=teams, label="Home Team"),
        gr.Dropdown(choices=teams, label="Away Team"),
        gr.Dropdown(choices=refs,  label="Referee"),
        gr.Dropdown(choices=times, label="Time of Day", value='Afternoon')
    ],
    outputs=[
        gr.Number(label="Predicted Total Yellows (raw)"),
        gr.Number(label="Predicted Total Yellows (rounded)")
    ],
    title="Premier League Total Yellows Predictor (Hoeffding Tree Regressor, online)",
    description=(
        "Select Home/Away teams, Referee and Time of Day. "
        "Odds-derived features use historical averages; other features come from recent matches. "
        "Model: Hoeffding Tree Regressor with polynomial feature expansion and walk-forward training. "
        "Log-transform is applied only if non-log predictions would be negative."
    )
)

iface.launch()
```
